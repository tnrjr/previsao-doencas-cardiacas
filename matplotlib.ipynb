```python
import numpy as np

```

```python
x = np.linspace(0, 5, 11)

x
```

```python
y = x ** 2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y, "red")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.title("Título")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")
plt.subplot(1, 2, 2)
plt.plot(y, x, "g*-")
```

# Orientação à objetos no matplotlib

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, "blue")
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Título")

```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, "blue")
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Título")

axes2.plot(y,x, "red")
axes.set_xlabel("Label x2")
axes.set_ylabel("Label y2")
axes.set_title("Título2")

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, "b--", label="x**2")
ax.plot(x, x**3, "g*-", label="x**3")

ax.set_title("Título")
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo y")

ax.legend(loc=6)
                  
                 
```
